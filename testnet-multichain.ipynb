```python
#Libraries
import pandas as pd
import numpy as np
```

## 1. Generate SQL scripts fromm datasets

```python
#Multichain activities overtime for testnet
df_t = pd.read_csv('Data Sources/testnet-multichain-analysis.csv')

#GENERATE QUERY TO CREATE A TABLE ON Flipside
query = []
for i,row in df_t.iterrows():
     print(f"SELECT {row.row_id}, '{row.chain_name}', {row.transactions},{row.tx_accumulated},{row.active_users},{row.new_users},{row.users_accumulated},{row.new_users_percent},{row.retention_ratio},{row.avg_gas_spent},{row.total_gas_spent}, {row.avg_gas_price} UNION ALL".replace('nan','null'))

query
```

```python
#Added more metrics to the summary table
avg_retention = df_t.groupby('chain_name')['retention_ratio'].mean()
avg_new_users = df_t.groupby('chain_name')['new_users'].mean()
display(avg_retention,avg_new_users)
```

```python
#Summary Table - Activities Accumulated
df_a = pd.read_csv('Data Sources/testnet-activtities-accumulated.csv')
df_a.avg_new_users_percent = round(df_a.avg_new_users_percent,3)
df_a
```

```python
df_a = pd.concat([df_a.set_index('chain_name'), avg_retention.to_frame(),avg_new_users.to_frame()], ignore_index= False, axis= 1)
df_a = df_a.reset_index().rename(columns = {'new_users': 'avg_new_users', 'retention_ratio':"avg_retention_ratio"})
df_a 
```

```python

for i,row in df_a.iterrows():
     print(f"SELECT '{row.chain_name}', {row.total_tx},{row.total_users},{row.avg_new_users},{row.avg_new_users_percent},{row.avg_retention_ratio},{row.avg_gas_spent},{row.avg_avg_gas_price} UNION ALL".replace('nan','null'))
```
